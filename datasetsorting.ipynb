```python
import os
import cv2
import numpy as np
import shutil
```

```python
#pick classes
my_list = []
while True:
    a=input('Enter class name or 0 to end:')
    if a!=str(0):
        my_list.append(a)
    else:
        break
print(my_list)
```

```python
#make dir
for clss in my_list:
    directory = clss
    #Parent_dir is directory where you want to save data
    parent_dir = input("Enter directry you want to save data")  
    path = os.path.join(parent_dir, directory)
    os.makedirs(path, exist_ok = True)
    print("Directory '%s' created successfully" % directory)
```

```python
#save images
vl=[]
#when asked input path to images you need to sort
imgpath=input('Paste file path to images:')
imgpath=(f'{imgpath}/')
images = os.listdir(imgpath)
inc=0
val=1
for cl in my_list:
    print(f'Enter {val} to save in {cl}')
    vl.append(val)
    val+=1
for z in range(len(images)):
    inc=0
    keypress=cv2.waitKey(inc)
    i=cv2.imread(imgpath+images[z])
    cv2.imshow('img',i)
    if keypress==ord(str(0)):
        cv2.destroyAllWindows()
        break
    elif keypress>=ord("1") and keypress<=(ord(str(len(my_list)))): 
        cv2.imwrite(f"{parent_dir}/{my_list[int(chr(keypress))-1]}/{images[z]}", cv2.imread(imgpath+images[z-1]))  
print('done')
cv2.waitKey() 
cv2.destroyAllWindows()
```
